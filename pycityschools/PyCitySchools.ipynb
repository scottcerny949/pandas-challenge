```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset. 
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Calculate the total number of schools
show_schools = school_data_complete["school_name"].unique()
show_schools
```

```python
num_schools = int(len(show_schools))
num_schools
```

```python
#Calculate the total number of students
num_students = int(len(school_data_complete["Student ID"]))
num_students
```

```python
#Calculate the total budget
total_budget = int(school_data_complete["budget"].unique().sum())
total_budget
```

```python
#Calculate the average math score
avg_math = round(school_data_complete["math_score"].mean(), 6)
avg_math
```

```python
#Calculate the average reading score
avg_read = round(school_data_complete["reading_score"].mean(), 6)
avg_read
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_pass = len(school_data_complete.loc[school_data_complete["math_score"] >= 70, :])
math_pass
math_total = len(school_data_complete["math_score"])
math_total
math_pass_pct = round((math_pass / math_total)*100, 6)
math_pass_pct
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
read_pass = len(school_data_complete.loc[school_data_complete["reading_score"] >= 70, :])
read_pass
read_total = len(school_data_complete["reading_score"])
read_total
read_pass_pct = round((read_pass / read_total)*100, 6)
read_pass_pct
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
both_pass = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
both_pass
both_pass_pct = round((len(both_pass) / num_students)*100, 6)
both_pass_pct
```

```python
#Create a dataframe to hold the above results
summary_df = pd.DataFrame({"Total Schools": [num_schools],
                           "Total Students": [num_students],
                           "Total Budget": [total_budget],
                           "Average Math Score": [avg_math],
                           "Average Reading Score": [avg_read],
                           "% Passing Math": [math_pass_pct],
                           "% Passing Reading": [read_pass_pct],
                           "% Overall Passing": [both_pass_pct]
                            })
summary_df
```

```python
#Optional: give the displayed data cleaner formatting
summary_df2 = pd.DataFrame({"Total Schools": [num_schools],
                           "Total Students": [num_students],
                           "Total Budget": [total_budget],
                           "Average Math Score": [avg_math],
                           "Average Reading Score": [avg_read],
                           "% Passing Math": str(math_pass_pct)+"%",
                           "% Passing Reading": str(read_pass_pct)+"%",
                           "% Overall Passing": str(both_pass_pct)+"%",
                            })
summary_df2.style.hide_index()
```

```python
#Set School Name as index
grouped_schools = school_data_complete.set_index('school_name').groupby(['school_name'])
grouped_schools.head()
```

```python
#School Type
school_type = school_data.set_index('school_name')['type']
school_type
```

```python
#Total Students
student_tot = school_data.set_index('school_name')['size']
student_tot
```

```python
#Total School Budget
school_budget = school_data.set_index('school_name')['budget']
school_budget
```

```python
#Per Student Budget
#budget / size for each school
student_budget = school_data.set_index('school_name')['budget'] / school_data.set_index('school_name')['size']
student_budget
```

```python
#Average Math Score
school_math = grouped_schools["math_score"].mean()
school_math
```

```python
#Average Reading Score
school_read = grouped_schools["reading_score"].mean()
school_read
```

```python
#% Passing Math
school_math_pass = school_data_complete[school_data_complete["math_score"] >= 70]
school_math_pass_2 = (school_math_pass.groupby("school_name")["math_score"].count()/student_tot)*100
school_math_pass_2
```

```python
#% Passing Reading
school_read_pass = school_data_complete[school_data_complete["reading_score"] >= 70]
school_read_pass_2 = (school_math_pass.groupby("school_name")["reading_score"].count()/student_tot)*100
school_read_pass_2
```

```python
#% Overall Passing (The percentage of students that passed math and reading.)
school_over_pass = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
school_over_pass_2 = (school_over_pass.groupby("school_name")["math_score"].count()/student_tot)*100
school_over_pass_2
```

```python
#Create a dataframe to hold the above results
school_summary = pd.DataFrame({
                             "School Type": school_type,
                             "Total Students": student_tot,
                             "Total School Budget": school_budget, 
                             "Per Student Budget": student_budget, 
                             "Average Math Score": school_math, 
                             "Average Reading Score": school_read, 
                             "% Passing Math": school_math_pass_2, 
                             "% Passing Reading": school_read_pass_2, 
                             "% Overall Passing": school_over_pass_2
                              })
school_summary
```

```python
#Sort and display the top five performing schools by % overall passing
sorted_over_pass = school_summary.sort_values("% Overall Passing", ascending = False)
sorted_over_pass.head()
```

```python
#Sort and display the five worst-performing schools by % overall passing.
bottom_over_pass = school_summary.sort_values("% Overall Passing")
bottom_over_pass.head()
```
